### PAnel Regression, Time series and ML

```python
# =======================================================================================
#  Definitive High-Impact Research Workflow: ESG Portfolios & Multivariate Analysis
# =======================================================================================
#
# Description: This script represents a best-practice framework for high-impact financial
#              research. It forms portfolios based on a composite ESG score, then employs
#              a primary multivariate regression (E+S+G) to isolate the impact of each
#              pillar. This is supported by univariate regressions, time series analysis
#              (GARCH, VAR), and a full suite of ML models (LASSO, DNN, LSTM) to provide
#              a comprehensive, publication-ready story.
#
# ---------------------------------------------------------------------------------------

# I. INITIAL SETUP AND LIBRARY IMPORTS
# ---------------------------------------------------------------------------------------
import pandas as pd
import numpy as np
import yfinance as yf
import statsmodels.api as sm
from statsmodels.tsa.stattools import grangercausalitytests
from statsmodels.tsa.api import VAR
import linearmodels as lm
from linearmodels.panel import PanelOLS
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LassoCV
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error
import warnings
from scipy.stats.mstats import winsorize
from arch import arch_model
import matplotlib.pyplot as plt
import seaborn as sns
import sys

# Deep Learning Libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# --- Script Settings ---
warnings.filterwarnings('ignore')
sns.set_style("whitegrid")
plt.rc('figure', figsize=(12, 6.5))
plt.rc('font', size=12)

# --- CONFIGURATION ---
START_DATE = "2019-01-01"
END_DATE = "2025-04-30"
ESG_DATA_PATH = '/content/daily_panel_esg_data_final.csv'
FF_DATA_PATH = '/content/Developed_5_Factors_Daily.csv'

# Column names from your CSV file
TICKER_COLUMN = 'Ticker'
DATE_COLUMN = 'Date'
ESG_SCORE_COLUMN = 'ESG'
E_PILLAR_COLUMN = 'E'
S_PILLAR_COLUMN = 'S'
G_PILLAR_COLUMN = 'G'


# II. DATA PREPARATION & PORTFOLIO FORMATION
# ------------------------------------------------------------------------------

print("--- Step 1: Data Loading and Merging ---")

# 1.1 Load ESG Panel Data
try:
    esg_df = pd.read_csv(ESG_DATA_PATH)
    column_mapping = {
        TICKER_COLUMN: 'ticker', DATE_COLUMN: 'Date', ESG_SCORE_COLUMN: 'ESG_Score',
        E_PILLAR_COLUMN: 'E_Score', S_PILLAR_COLUMN: 'S_Score', G_PILLAR_COLUMN: 'G_Score'
    }
    for source_col in column_mapping.keys():
        if source_col not in esg_df.columns:
            print(f"\nFATAL ERROR: Required column '{source_col}' not found in {ESG_DATA_PATH}.")
            sys.exit()
    esg_df.rename(columns=column_mapping, inplace=True)
    esg_df['Date'] = pd.to_datetime(esg_df['Date'])
    RAW_MATERIALS_TICKERS = esg_df['ticker'].unique().tolist()
except FileNotFoundError:
    print(f"FATAL ERROR: ESG data file not found at {ESG_DATA_PATH}."), sys.exit()

# 1.2 Load Fama-French Data and Stock Prices
ff_df = pd.read_csv(FF_DATA_PATH, parse_dates=['Date'])
ff_df.rename(columns={'Mom   ': 'WML', 'MOM': 'WML'}, inplace=True, errors='ignore')
ff_cols = ['Mkt-RF', 'SMB', 'HML', 'RMW', 'CMA', 'WML', 'RF']
ff_factors_only = [col for col in ff_cols if col in ff_df.columns and col != 'RF']
print(f"Using Fama-French factors: {ff_factors_only}")
stock_data = yf.download(RAW_MATERIALS_TICKERS, start=START_DATE, end=END_DATE, group_by='ticker', auto_adjust=True, progress=False)

# 1.3 Combine and Clean Data
all_data = []
for ticker in RAW_MATERIALS_TICKERS:
    try:
        if len(RAW_MATERIALS_TICKERS) > 1 and ticker not in stock_data.columns.get_level_values(0): continue
        stock_df = stock_data[ticker].copy() if len(RAW_MATERIALS_TICKERS) > 1 else stock_data.copy()
        if stock_df.empty or 'Close' not in stock_df.columns: continue
        stock_df.reset_index(inplace=True)
        stock_df['ticker'] = ticker
        stock_df['Return'] = stock_df['Close'].pct_change()
        merged = pd.merge(stock_df[['Date', 'ticker', 'Return']], esg_df, on=['Date', 'ticker'], how='left')
        final_merged = pd.merge(merged, ff_df, on='Date', how='left')
        all_data.append(final_merged)
    except Exception as e: print(f"Could not process ticker {ticker}: {e}")

final_df = pd.concat(all_data, ignore_index=True)
if 'RF' in final_df.columns: final_df['Excess_Return'] = final_df['Return'] - final_df['RF']
else: final_df['Excess_Return'] = final_df['Return']

# 1.4 Handle Missing Data and Outliers
print("\n--- Handling Missing Data and Outliers ---")
final_df.sort_values(by=['ticker', 'Date'], inplace=True)
esg_pillar_cols = ['ESG_Score', 'E_Score', 'S_Score', 'G_Score']
final_df[esg_pillar_cols] = final_df.groupby('ticker')[esg_pillar_cols].ffill()
for col in ['Excess_Return'] + ff_factors_only:
    final_df.loc[final_df[col].notna(), col] = winsorize(final_df.loc[final_df[col].notna(), col], limits=[0.01, 0.01])

# 1.5 ESG PORTFOLIO FORMATION
print("\n--- Step 2: Forming ESG Portfolios ---")
avg_esg = final_df.groupby('ticker')['ESG_Score'].mean().dropna()
if avg_esg.empty: print("\nFATAL ERROR: Cannot form ESG portfolios."), sys.exit()

esg_terciles = pd.qcut(avg_esg, 3, labels=['Low ESG', 'Medium ESG', 'High ESG'])
final_df['esg_portfolio'] = final_df['ticker'].map(esg_terciles)
print("Companies assigned to ESG portfolios:")
for portfolio_name in ['Low ESG', 'Medium ESG', 'High ESG']:
    tickers_in_portfolio = sorted(esg_terciles[esg_terciles == portfolio_name].index.tolist())
    print(f"  - {portfolio_name} Portfolio ({len(tickers_in_portfolio)} companies): {', '.join(tickers_in_portfolio)}")

model_df = final_df.set_index(['ticker', 'Date'])


# III. ESG PORTFOLIO ANALYSIS
# ------------------------------------------------------------------------------
for portfolio in ['Low ESG', 'Medium ESG', 'High ESG']:
    print(f"\n\n{'='*70}\nANALYSIS FOR: {portfolio.upper()} PORTFOLIO\n{'='*70}")
    portfolio_df = model_df[model_df['esg_portfolio'] == portfolio].copy().dropna(subset=['Excess_Return'])
    if portfolio_df.empty: continue

    # --- A. PANEL DATA ANALYSIS: From Broad to Specific ---
    print("\n--- A. Panel Data Analysis ---")

    # 1. MAIN ANALYSIS: Multivariate Regression (The "Competitive Test")
    print("\n*** 1. Main Fixed Effects Model (E, S, and G Pillars Combined) ***")
    multivariate_vars = ['E_Score', 'S_Score', 'G_Score'] + ff_factors_only
    panel_data = portfolio_df[multivariate_vars + ['Excess_Return']].dropna()
    if panel_data.empty or len(panel_data.index.get_level_values('ticker').unique()) < 2:
        print("Not enough data for multivariate panel analysis.")
    else:
        y_multi, X_multi = panel_data['Excess_Return'], sm.add_constant(panel_data.drop(columns='Excess_Return'))
        fe_model = PanelOLS(y_multi, X_multi, entity_effects=True).fit(cov_type='clustered', cluster_entity=True)
        print("This model tests which pillar has the most explanatory power when controlling for the others.")
        print(fe_model)

    # 2. SUPPORTING ANALYSIS: Univariate Regressions for Robustness
    print("\n*** 2. Supporting Fixed Effects Models (Pillars Tested Individually) ***")
    for component, score_col in {'Composite': 'ESG_Score', 'Environment': 'E_Score', 'Social': 'S_Score', 'Governance': 'G_Score'}.items():
        panel_data_uni = portfolio_df[[score_col] + ff_factors_only + ['Excess_Return']].dropna()
        if not panel_data_uni.empty and len(panel_data_uni.index.get_level_values('ticker').unique()) > 1:
            y_uni, X_uni = panel_data_uni['Excess_Return'], sm.add_constant(panel_data_uni.drop(columns='Excess_Return'))
            fe_model_uni = PanelOLS(y_uni, X_uni, entity_effects=True).fit(cov_type='clustered', cluster_entity=True)
            print(f"\n--- {component} Pillar Only ---\n{fe_model_uni}")

    # --- B. PORTFOLIO-LEVEL TIME SERIES ANALYSIS ---
    print("\n--- B. Portfolio-Level Time Series Analysis ---")
    portfolio_ts = portfolio_df.reset_index().groupby('Date').mean(numeric_only=True)
    if len(portfolio_ts) > 30:
        garch = arch_model(portfolio_ts['Excess_Return'].dropna() * 100, vol='Garch', p=1, q=1).fit(disp='off')
        print(f"\n*** GARCH Volatility Persistence (Alpha+Beta): {garch.params['alpha[1]'] + garch.params['beta[1]']:.4f} ***")

        gc_pval = grangercausalitytests(portfolio_ts[['Excess_Return', 'ESG_Score']].dropna(), [3], verbose=False)[3][0]['ssr_ftest'][1]
        print(f"*** Granger Causality (ESG -> Returns) p-value: {gc_pval:.4f} ***")
        if gc_pval < 0.1:
            try:
                irf = VAR(portfolio_ts[['Excess_Return', 'ESG_Score']].dropna().diff().dropna()).fit(10, 'aic').irf(10)
                irf.plot(orth=False, impulse='ESG_Score', response='Excess_Return')
                plt.suptitle(f'Impact of ESG Shock on Returns - {portfolio} Portfolio', fontsize=16)
                plt.show()
            except Exception as e: print(f"Could not run VAR/IRF: {e}")

    # --- C. MACHINE LEARNING FORECASTING & INTERPRETATION ---
    print("\n--- C. Advanced Machine Learning & Visualization ---")
    ml_df = portfolio_ts.copy()
    ml_df['Return_Lag1'] = ml_df['Excess_Return'].shift(1)
    ml_df.dropna(inplace=True)

    if len(ml_df) < 50: continue

    features = [c for c in esg_pillar_cols + ff_factors_only + ['Return_Lag1'] if c in ml_df.columns]

    # ### FIX: Define the 'target' variable ###
    target = 'Excess_Return'

    X_ml, y_ml = ml_df[features], ml_df[target]
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_ml)
    train_size = int(len(X_scaled) * 0.8)
    X_train, X_test = X_scaled[:train_size], X_scaled[train_size:]
    y_train, y_test = y_ml.iloc[:train_size], y_ml.iloc[train_size:]

    # Model Training
    tf.random.set_seed(42)
    lasso = LassoCV(cv=TimeSeriesSplit(n_splits=5)).fit(X_train, y_train)
    dnn = Sequential([Dense(64, 'relu', input_shape=(X_train.shape[1],)), Dropout(0.2), Dense(32, 'relu'), Dense(1)])
    dnn.compile('adam', 'mse')
    dnn.fit(X_train, y_train, epochs=100, validation_split=0.2, callbacks=[EarlyStopping('val_loss', patience=10)], verbose=0)
    lstm = Sequential([LSTM(50, 'relu', input_shape=(1, X_train.shape[1])), Dropout(0.2), Dense(1)])
    lstm.compile('adam', 'mse')
    lstm.fit(X_train[:, None, :], y_train, epochs=100, validation_split=0.2, callbacks=[EarlyStopping('val_loss', patience=10)], verbose=0)

    # Reporting & Visualization
    print("\n--- Model Performance (RMSE on Test Set) ---")
    print(f"LASSO RMSE: {np.sqrt(mean_squared_error(y_test, lasso.predict(X_test))):.6f}")
    print(f"DNN RMSE:   {np.sqrt(mean_squared_error(y_test, dnn.predict(X_test))):.6f}")
    print(f"LSTM RMSE:  {np.sqrt(mean_squared_error(y_test, lstm.predict(X_test[:, None, :]))):.6f}")

    # Feature Importance Plot
    lasso_coeffs = pd.Series(lasso.coef_, index=features).sort_values()
    plt.figure()
    lasso_coeffs[lasso_coeffs != 0].plot(kind='barh', color=np.where(lasso_coeffs[lasso_coeffs != 0] > 0, 'seagreen', 'indianred'))
    plt.title(f'Feature Importance for {portfolio} Portfolio (LASSO)')
    plt.xlabel('Coefficient Value (Impact on Return)')
    plt.tight_layout()
    plt.show()
```
