## Python Code for Content and Sentiment Analysis

### Steps Explained:
#### Preprocessing and Tokenization: Text responses are tokenized, converted to lowercase, and cleaned of stopwords.

```python

import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt

# Download NLTK resources
##nltk.download('punkt')
#nltk.download('stopwords')

# Load the survey data
data = pd.read_csv('survey_data.csv')

print(data.columns)
# Import necessary libraries
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set a custom download directory and download NLTK resources
nltk_data_dir = os.path.expanduser("~/nltk_data")
os.makedirs(nltk_data_dir, exist_ok=True)
nltk.data.path.append(nltk_data_dir)

nltk.download('punkt', quiet=True, download_dir=nltk_data_dir)
nltk.download('stopwords', quiet=True, download_dir=nltk_data_dir)

# Load the survey data
data = pd.read_csv('survey_data.csv')

# Verify the structure of the dataset
print(data.head())

# ---- DESCRIPTIVE ANALYSIS ----
print("\nDescriptive Analysis")
total_responses = len(data)
print(f"Total Responses: {total_responses}")

if 'Response' not in data.columns:
    raise ValueError("The 'Response' column is missing from the dataset.")

if data['Response'].isnull().any():
    print("Warning: Missing values found in 'Response' column. Handling them with fillna('').")

# Ensure 'Response' column has no missing values
data['Response'] = data['Response'].fillna('')

# Summary statistics for response lengths
data['Response_Length'] = data['Response'].apply(lambda x: len(str(x).split()))
print("Response Length Statistics:")
print(data['Response_Length'].describe())

# Plot response lengths
plt.figure(figsize=(8, 5))
sns.histplot(data['Response_Length'], kde=True, bins=20, color='blue')
plt.title('Distribution of Response Lengths')
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.show()

# ---- CONTENT ANALYSIS ----
print("\nContent Analysis")

# Tokenize and preprocess responses
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # Simple tokenization using string split to avoid potential issues with word_tokenize
    tokens = str(text).lower().split()
    return [word for word in tokens if word.isalpha() and word not in stop_words]

data['Tokens'] = data['Response'].apply(preprocess_text)

# Flatten list of all tokens
all_tokens = [token for tokens in data['Tokens'] for token in tokens]

# Most common words
word_freq = Counter(all_tokens)
print("Most Common Words:")
print(word_freq.most_common(10))

# Plot most common words
common_words = word_freq.most_common(10)
words, counts = zip(*common_words)
plt.figure(figsize=(8, 5))
sns.barplot(x=list(words), y=list(counts), palette='viridis')
plt.title('Top 10 Most Common Words')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()

# ---- SENTIMENT ANALYSIS ----
print("\nSentiment Analysis")

# Initialize sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Compute sentiment scores
data['Sentiment_Scores'] = data['Response'].apply(lambda x: analyzer.polarity_scores(str(x)))
data['Compound_Score'] = data['Sentiment_Scores'].apply(lambda x: x['compound'])
data['Sentiment'] = data['Compound_Score'].apply(
    lambda x: 'Positive' if x > 0.05 else ('Negative' if x < -0.05 else 'Neutral')
)

# Sentiment distribution
print("Sentiment Distribution:")
print(data['Sentiment'].value_counts())

# Plot sentiment distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Sentiment', data=data, palette='pastel')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()

# Average sentiment score
average_sentiment = data['Compound_Score'].mean()
print(f"Average Compound Sentiment Score: {average_sentiment:.2f}")

# Save the enriched dataset with analysis results
data.to_csv('survey_data_analysis.csv', index=False)
print("\nAnalysis complete. Results saved to 'survey_data_analysis.csv'")
```
