#### To download historical stock prices for the specified companies from 2014 to 2023, you can use the yfinance library in Python, which allows easy access to Yahoo Finance data.

## Below is a Python script that retrieves the stock data for the listed companies:

# Output:
### The script will download the historical stock prices for the specified companies from 2014 to 2023 and save the data to a CSV file named historical_stock_prices_2014_2023.csv.


```python
import yfinance as yf
import pandas as pd

# List of companies with their stock symbols
companies = {
    "PayPal Holdings, Inc.": "PYPL",
    "Broadcom Inc.": "AVGO",
    "The Walt Disney Company": "DIS",
    "Taiwan Semiconductor Manufacturing Company Limited": "TSM",
    "Alibaba Group Holding Limited": "BABA",
    "Advanced Micro Devices, Inc.": "AMD",
    "Adobe Inc.": "ADBE",
    "Apple Inc.": "AAPL",
    "Alphabet Inc.": "GOOG",
    "Meta Platforms, Inc.": "META"
}

# Define start and end date for the data retrieval
start_date = "2014-01-01"
end_date = "2023-12-31"

# Create a dictionary to hold data for each company
stock_data = {}

# Fetch stock data for each company
for name, symbol in companies.items():
    print(f"Downloading data for {name} ({symbol})...")
    stock_data[symbol] = yf.download(symbol, start=start_date, end=end_date)

# Combine all data into a single DataFrame
combined_data = pd.concat(stock_data.values(), keys=stock_data.keys())

# Save to CSV
combined_data.to_csv('historical_stock_prices_2014_2023.csv')

print("Data download complete. Historical stock prices saved to 'historical_stock_prices_2014_2023.csv'.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
from semopy import Model

# Load the data
data = pd.read_csv("/content/Metaverse_ESG_20.csv")

# List of companies based on column names
companies = ['IBM', 'SIE', 'BMW', 'ADS', 'ADSK', 'NKE', 'NVDA', 'BABA',
             'ADBE', 'AAPL', 'AVGO', 'PYPL', 'MSFT', 'META', 'AMZN']

# Initialize a directed graph
G = nx.DiGraph()

for company in companies:
    try:
        # Select relevant columns for the current company
        company_data = data.filter(like=company)
        company_data = company_data.dropna()

        if company_data.empty:
            print(f"No sufficient data for {company}, skipping...")
            continue

        variables = company_data.columns.tolist()
        print(f"Available variables for {company}: {variables}")

        # Construct the model dynamically based on available data
        measurement_model = []
        if f'{company}_Env_Risk' in variables:
            measurement_model.append(f"E_Score =~ {company}_Env_Risk")
        if f'{company}_Soc_Risk' in variables:
            measurement_model.append(f"S_Score =~ {company}_Soc_Risk")
        if f'{company}_Gov_Risk' in variables:
            measurement_model.append(f"G_Score =~ {company}_Gov_Risk")

        # Add nodes and edges to the graph
        for line in measurement_model:
            latent_variable, _, observed_variable = line.partition(' =~ ')
            G.add_node(latent_variable.strip())
            G.add_node(observed_variable.strip())
            G.add_edge(latent_variable.strip(), observed_variable.strip())

    except KeyError as e:
        print(f"Missing data for {company}: {e}")
        continue

# Plotting the combined graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, arrows=True, node_size=3000, node_color='lightblue', font_size=10, font_weight='bold')
plt.title("Combined Path Diagram for All Companies")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from semopy import Model

# Load the data
data = pd.read_csv("/content/Metaverse_ESG_20.csv")

# Check the structure of the dataset
print(data.head())
print(data.columns)

# Define a simplified SEM model
model_desc = """
# Measurement model
E_Score =~ LOW_Env_Risk + LOW_Soc_Risk + LOW_Gov_Risk
S_Score =~ ANSS_Env_Risk + ANSS_Soc_Risk + ANSS_Gov_Risk
G_Score =~ PTC_Env_Risk + PTC_Soc_Risk + PTC_Gov_Risk
"""

# List of companies based on column names
companies = ['IBM', 'SIE', 'BMW', 'ADS', 'ADSK', 'NKE', 'NVDA', 'BABA',
             'ADBE', 'AAPL', 'AVGO', 'PYPL', 'MSFT', 'META', 'AMZN']  # Adjust as needed

# Initialize a list to store the results
results = []

for company in companies:
    # Create a DataFrame for the current company
    try:
        company_data = data[[f'{company}_ESG', f'{company}_Env_Risk', f'{company}_Soc_Risk', f'{company}_Gov_Risk']]
    except KeyError as e:
        print(f"Missing data for {company}: {e}")
        continue

    # Check for null values and drop rows with any nulls
    company_data = company_data.dropna()

    # Check if there is sufficient data for the model fitting
    if company_data.empty:
        print(f"No sufficient data for {company}, skipping...")
        continue

    # Fit the model
    model = Model(model_desc)
    try:
        model.fit(company_data)
    except Exception as e:
        print(f"Error fitting model for {company}: {e}")
        continue

    # Store the results
    results.append(model)

    # Plotting the path diagram
    fig, ax = plt.subplots(figsize=(8, 6))
    semopy.plot(model, ax=ax)
    plt.title(f"Path Diagram for {company}")
    plt.show()
```
